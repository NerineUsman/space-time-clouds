```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
N  = 100
image = xr.open_dataset('../data/start_image0.nc').isel(i = slice(100), j = slice(100))
image.ct.plot()
image
```

```python
pixels = image.stack(pixel=("i", "j"))
pixels = pixels.where(pixels.ct == 10, drop = True)

plt.scatter(pixels.i, pixels.j)
pixels
```

```python
pixels = pixels.sortby('i')

pixels['cloud_number'] = (['pixel'], np.zeros(pixels.sizes['pixel']))

cloud_number = 1
while (pixels.cloud_number == 0).sum() > 0:
    pixel = pixels.cloud_number.argmin()
    
    
    pixels.cloud_number[pixel] = cloud_number



    nb_pixels = pixels.where((np.abs(pixels.i - pixels.i.isel(pixel = pixel)) < 6e3) &
                             (np.abs(pixels.j - pixels.j.isel(pixel = pixel)) < 6e3), drop = True)

    if nb_pixels.sizes['pixel'] == 0:
        cloud_number += 1
        print(cloud_number)
        
    else:
        pixels.cloud_number.loc[dict(pixel = nb_pixels.pixel)] = nb_pixels.cloud_number.max()
        cloud_number += 1
        


    

# pixel = 
pixels
# pixel
# clusters['centroids_i'] = (['cluster'], np.array[])
# clusters['centroids_j'] = (['cluster'], np.array[])
```
