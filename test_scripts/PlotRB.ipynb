```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace( 0 , 1, 51 )
y = np.linspace( 0, 1, 51)
X, Y  = np.meshgrid(x, y)
X, Y = X.flatten(), Y.flatten()
hue_c = .9
rgb_cycle = np.vstack((            # Three sinusoids
    (Y >= X) * 1 +  (Y < X) * (Y - X + 1) + ((X + Y) > 1) * -(X + Y -1) * hue_c, # scaled to [0,1]
    ((X + Y) < 1) * ( -X - Y + 1) * hue_c  , # 120 phase shifted.
    (Y >= X) * (X - Y + 1) +  (Y < X) * 1 + ((X + Y) > 1) * -(X + Y -1) * hue_c )).T # Shape = (60,3)

rgb_cycle = np.where(rgb_cycle<0, 0, rgb_cycle)
rgb_cycle
plt.scatter(X, Y, s = 15, c = rgb_cycle)
plt.colorbar()
# plt.show()

# rgb_cycle.min()
```

```python
c = .92
rgb_cycle = np.vstack((            # Three sinusoids
    (-X + 1) * c, # scaled to [0,1]
    (-Y + 1) * c  , # 120 phase shifted.
    (- X - Y + 1))).T # Shape = (60,3)
rgb_cycle = np.roll(rgb_cycle, 2, axis = 1)
rgb_cycle = np.where(rgb_cycle<0, 0, rgb_cycle)
rgb_cycle
plt.scatter(X, Y, s = 15, c = rgb_cycle)
plt.colorbar()
# plt.show()

# rgb_cycle.min()
```
